```python
import numpy as np; 
import matplotlib.pyplot as plt; 
import scipy.stats as st
```

```python
#4.1.1
N=2000
x=st.binom.rvs(1,0.5,size = N)
print(x)
```

```python
#4.1.2
# Y_n converge vers E(X_1)=p=0.5 avec probabilite 1
#(cette convergence est dite presque sure)
```

```python
#4.1.3
y = np.cumsum(x) / np.arange(1, N + 1)
print(y)

y1 = np.cumsum(x)/(1+np.arange(N))
print(y1)
#solution 1
Y=[np.mean(x[:i]) for i in range(1,N+1)]
#ou
Y1=[np.mean(x[:(n+1)]) for n in range(N)]
```

```python
#4.1.4
plt.plot(np.arange(1,N+1),Y)
plt.plot(np.arange(1,N+1),np.repeat(0.5,N),color='red')
plt.title("Illustation de la LGN B(1/2)")
plt.show()
```

```python
#4.1.5
N=2000
n = st.norm.rvs(1,2,size = N) #Attention donner ecart-type 2! 
plt.hist(n, bins=20, density=True)
print(n)
n1 = np.linspace(-9,9,901)
d = st.norm.pdf(n1,1,2)
plt.plot(n1,d)
plt.title("Simulation et denstite pour N(1,4)")
```

```python
#4.1.6
yn = np.cumsum(n)/(1+np.arange(N))
print(yn)
plt.plot(yn)
plt.plot(np.repeat(1,N),color='red')
plt.title("Illustration de la LGN pour N(1,4)")
```

```python
#4.2.1
n = 100
N = 1000
X = st.expon.rvs(size=(N,n), loc=0, scale = 1)
print(X)

xbar1 =[np.mean(X[i,0]) for i in range(N)]
xbar3 =[np.mean(X[i,0:3]) for i in range(N)]
xbar100 =[np.mean(X[i,0:100]) for i in range(N)]

V1 = np.asarray(xbar1) - 1.0
V3 = np.sqrt(3)*(np.asarray(xbar3) - 1.0)
V100 = np.sqrt(100)*(np.asarray(xbar100) - 1.0)
```

```python
fig, ax = plt.subplots(1,3)
ax[0].hist(V1)
ax[1].hist(V3)
ax[2].hist(V100)
```
